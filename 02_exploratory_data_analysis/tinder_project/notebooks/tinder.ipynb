### Dropping columns with too many missing values
Several columns contain more than 70–80% missing values, especially follow-up variables collected weeks after the event.  
They are not needed for the analysis, which focuses only on what happens during the speed-dating session itself.

Can people accurately predict their own perceived value in the dating market?

```python
self_cols = [
    "self_attr",
    "self_sinc",
    "self_intel",
    "self_fun",
    "self_amb"
]

partner_cols = [
    "o_rating_attr",
    "o_rating_sinc",
    "o_rating_intel",
    "o_rating_fun",
    "o_rating_amb"
]

cols_self_patner = self_cols + partner_cols
df_self_partner = df_core[cols_self_patner]

df_self_partner.head()
```

```python
self_mean = df_self_partner[self_cols].mean().round(2)
partner_mean = df_self_partner[partner_cols].mean().round(2)

df_self_vs_partner = pd.DataFrame({
    "attribute": ["Attractiveness", "Sincerity", "Intelligence", "Fun", "Ambition"],
    "self_rating": self_mean.values,
    "partner_rating": partner_mean.values
})

df_self_vs_partner
```

```python
df_long = df_self_vs_partner.melt(
    id_vars="attribute",
    value_vars=["self_rating", "partner_rating"],
    var_name="rating_type",
    value_name="score"
)

df_long
```

```python
px.bar(
    df_long,
    x="attribute",
    y="score",
    color="rating_type",
    barmode="group",
    title="Self-Rating vs Partner-Rating Across Traits"
)
```

People are not very accurate in predicting their own perceived value in the dating market.
On average, participants give themselves higher ratings than their partners give them, with a consistent gap across all traits.

To understand whether people can truly predict how others will perceive them, we also need to measure the relationship between self-ratings and partner-ratings.

```python
attributes = ["attr", "sinc", "intel", "fun", "amb"]

correlation_results = []

for attribute in attributes:
    self_col = f"self_{attribute}"
    partner_col = f"o_rating_{attribute}"

    corr_value = df_core[self_col].corr(df_core[partner_col])

    correlation_results.append({
        "attribute": attribute,
        "correlation": round(corr_value, 3)
    })

df_corr = pd.DataFrame(correlation_results)
df_corr
```

All correlations between self-ratings and partner-ratings are extremely low.
A perfect prediction would show a correlation close to 1, but the values observed here range from 0.00 to 0.18, which indicates almost no relationship between how people see themselves and how their dates actually perceive them.

People are not accurate at predicting their own perceived value in the dating market.
Self-ratings and partner-ratings correlate extremely weakly (0.00–0.18), showing that how participants evaluate themselves has almost no relationship with how others evaluate them during the speed dates.
Combined with the systematic self-overestimation observed earlier, these results show that people have a poor intuition about how desirable they actually are.

In terms of getting a second date, is it better to be someone's first speed date of the night or their last?

```python
cols_to_keep = ["date_order", "partner_decision"]
df_order = df_core[cols_to_keep]
df_order
```

```python
df_order = (
    df_order
    .groupby("date_order")
    .mean()
    .round(2)
    .reset_index()
    .rename(columns={"partner_decision": "yes_rate"})
)
df_order
```

```python
px.line(
    df_order,
    x="date_order",
    y="yes_rate",
    markers=True,
    title="Yes rate by date order"
)
```

Participants are more likely to receive a “yes” at the beginning of the evening than at the end.
The first position shows the highest yes-rate, the middle positions remain relatively stable with moderate values, and although some late positions rise again, the final ones tend to show the lowest rates.
This suggests that the order in which someone is met can influence the outcome: early encounters generally perform better than later ones.

Overall, several factors shape whether participants want a second date.  
First, men and women differ in the attributes they say they value, but in practice, attractiveness and shared interests are the strongest real predictors of a “yes”. Shared race and religion, although rated as somewhat important by a few participants, show almost no measurable impact on the final decision.  
Participants also tend to overestimate their own desirability: self-ratings are consistently higher than partner ratings, and the correlation between the two is extremely low, indicating that people are not accurate in predicting how they will be perceived.  
The order of the encounters also matters: early positions in the evening receive the highest yes-rates, middle positions remain stable, and the very last positions tend to receive the lowest ones.  
Taken together, the results show that immediate impressions such as appearance, perceived connection, and position in the sequence play a far greater role in speed-dating outcomes than demographic similarities or self-perception.
